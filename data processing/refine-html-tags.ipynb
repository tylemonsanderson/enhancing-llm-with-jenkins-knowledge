```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
file_path = 'Community Questions.csv'

df = pd.read_csv(file_path)

df.head(1)
```

```python
# print(df['Question Body'][0])
```

```python
df_updated = df.copy()
```

```python
def remove_html_tags(text):
    if pd.isna(text):  
        return text

    soup = BeautifulSoup(text, 'html.parser')

    # Find all tags except <code>
    for tag in soup.find_all(True):
        if tag.name != 'code':
            tag.unwrap()  # Remove the tag, keep its contents

    return str(soup)

df_updated['questions'] = df_updated['questions'].apply(remove_html_tags)
df_updated['answers'] = df_updated['answers'].apply(remove_html_tags)
```
